The line plot above does not provide useful information, this is because all the points in the dataset are plotted to the line

```python
bins_e = np.arange(2005, 2014+1, 1)
bins_c = bins_e[:-1] + 1

year_binned = pd.cut(df2['Year'], bins_e, include_lowest = True)
rate_mean = df2['BorrowerRate'].groupby(year_binned).mean()

plt.errorbar(x = bins_c, y = rate_mean);
```

This is a better plot and we can see that there was a rising theme in the average borrower rate from 2007 till 2011 but the rate has been dropping since then and is at an all time low in 2014

The relationships in the bivariate exploration were somewhat expected. It makes sense that people in higher income ranges will be allowed to borrow larger sums of money and that the lower the propser rating given to an individual the higher their borrower rate would be. Perhaps the most surprising thing was that there was almost no correlation between the debt to income ratio and the borrower rate and APR as I expected there to be one.

## Multivariate Exploration

Here I want to explore the relationship between the income range and the borrower rate for each year in the dataset and also based on if the borrower is a homeowner or not.

```python
def mean_poly(x, y, bins = 10, **kwargs):
    """ Custom adapted line plot code. """
    # set bin edges if none or int specified
    if type(bins) == int:
        bins = np.linspace(x.min(), x.max(), bins+1)
    bin_centers = (bin_edges[1:] + bin_edges[:-1]) / 2

    # compute counts
    data_bins = pd.cut(x, bins, right = False,
                       include_lowest = True)
    means = y.groupby(data_bins).mean()

    # create plot
    plt.errorbar(x = bin_centers, y = means, **kwargs)

bin_edges = np.arange(2005, 2014+1, 1)
g = sb.FacetGrid(data = df2, hue = 'IncomeRange', height = 5)
g.map(mean_poly, "Year", "BorrowerRate", bins = bin_edges)
g.set_ylabels('mean(BorrowerRate)')
g.add_legend()
```

From the plot above the relationships between the income range and borrower rate follow a very similar trend between the years, but we can also see that the average rates are lower for those in higher income groups than for those in the lower income or not employed groups. Individuals who do not have their income range displayed were from earlier accounts when the company probably did not put people into income ranges

```python
sb.pointplot(data = df2, x = 'IncomeRange', y = 'BorrowerRate', hue = 'IsBorrowerHomeowner');
plt.xticks(rotation=30);
```

Here we can see that borrowers who own homes on average have a lower borrower rate and it follows from the plot above it that higher income ranges have lower borrower rates.
