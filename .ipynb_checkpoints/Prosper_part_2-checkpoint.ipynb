# Effects of Income Ranges and Company Ratings on loans at Prosper 


# Investigation Overview

In this investigation, I wanted to focus on the relationships Income Range and the Prosper Rating had with other factors in the dataset, especially the borrower rate

# Dataset Overview

The Data consisted of data of 113,937 loans ranging from 2005-2014 with 81 different variables for each loan. This investigation focused on only 14 of those variables. A high Prosper Rating means an individual is safer or less likely to default on a loan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv('prosperLoanData.csv')
```

```python
df["DebtToIncomeRatio"].fillna(df["MonthlyLoanPayment"] / (df["StatedMonthlyIncome"] + 1), inplace = True)
```

```python
df2 = df[['ListingCreationDate', 'LoanStatus', 'CreditGrade','ProsperRating (numeric)', 'ProsperRating (Alpha)', 'ProsperScore', 'ListingCategory (numeric)', 'IsBorrowerHomeowner', 'CreditScoreRangeUpper', 'IncomeRange', 'LoanOriginalAmount',  'BorrowerRate', 'BorrowerAPR', 'DebtToIncomeRatio']]
```

```python
income_ranges = ['Not employed', '$0', '$1-24,999', '$25,000-49,999', '$50,000-74,999', '$75,000-99,999', '$100,000+', 'Not displayed']
incomerange = pd.api.types.CategoricalDtype(ordered = True, categories = income_ranges)
df2['IncomeRange'] = df2['IncomeRange'].astype(incomerange)
```

```python
df2['ProsperRating (Alpha)'].fillna(df['CreditGrade'], inplace = True)
```

```python
df2 = df2.replace({'ProsperRating (Alpha)': {'HR': 1, 'E': 2, 'D': 3, 'C': 4, 'B': 5, 'A': 6, 'AA': 7, }})
```

```python
df2.dropna(subset=['ProsperRating (Alpha)'], inplace=True)
```

```python
# Drop rows with NC
df2 = df2[df2['ProsperRating (Alpha)'] != 'NC']
```

```python
df2.dropna(subset=['ProsperRating (Alpha)'], inplace=True)
```

```python
df2['ProsperRating (Alpha)'] = df2['ProsperRating (Alpha)'].astype(int)
```

```python
df2.drop(['CreditGrade', 'ProsperRating (numeric)'], axis = 1, inplace = True)
```

```python
df2 = df2.rename(columns={'ProsperRating (Alpha)': 'ProsperRating'})
```

```python
df2['ListingCreationDate'] = pd.to_datetime(df2['ListingCreationDate'])
```

```python
df2['Year'] = pd.DatetimeIndex(df2['ListingCreationDate']).year
```

```python
num_vars =['ProsperRating', 'ProsperScore', 'ListingCategory (numeric)', 'CreditScoreRangeUpper', 'LoanOriginalAmount', 'BorrowerRate', 'BorrowerAPR', 'DebtToIncomeRatio', 'Year']
```

```python
df2.BorrowerAPR.isnull().sum()
```

```python
df2.dropna(subset=['BorrowerAPR'], inplace=True)
```

```python
df2.BorrowerAPR.isnull().sum()
```

# Distribution of Income Range

Most of the individuals in the Dataset were located between the income ranges of 25,000 Dollars to 74,999 Dollars. There were a minimal number of people stated as unemployed or earning no income. There were also 7600 people who did not have their income range displayed

```python
base_color = sb.color_palette()[0]
sb.countplot(data = df2, x = 'IncomeRange', color=base_color);
plt.xticks(rotation=80);
plt.title('Number of People in each Income Range Group')
plt.xlabel('Income Range groups');
```

# Distribution of Borrower Rate and APR

Both the Borrower Rate and the Borrower APR have a similar distribution that range from 0 t0 0.5. This is due to the fact that Borrower APR is basically the Borrower Rate plus a few additional fees from the loan company

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
bin_edges = np.arange(0, df2.BorrowerRate.max()+0.01, 0.01);
plt.hist(data = df2, x = 'BorrowerRate', bins = bin_edges);
plt.title('Borrower Rate Count')
plt.xlabel('Values')

plt.subplot(1, 2, 2)
bin_edges = np.arange(0, df2.BorrowerAPR.max()+0.01, 0.01);
plt.hist(data = df2, x = 'BorrowerAPR', bins = bin_edges);
plt.xlabel('Values')

plt.subplots_adjust(bottom=0.15, wspace=0.3)
plt.title('Borrower APR Count');
```

# Income Range vs Loan Original Amount

Most of the Loan Original Amounts in each income range are below 5000 for income ranges below 75,000 Dollars. The distribution is more spread out for income ranges over 75,000. The maximum loan amount is 35,000 and this was given to an individual in the income range of over 100,000 Dollars

```python
plt.figure(figsize=(12,5))
sb.violinplot(data = df2, x = 'IncomeRange', y = 'LoanOriginalAmount', color=base_color, inner=None);
plt.xticks(rotation=30);
plt.title('Original Loan Amounts per Income Range')
plt.ylabel('Original Loan Amount');
plt.xlabel('Income Range Groups');
```

# Prosper Rating vs Borrower Rate

There is a negative correlation between the Prosper Rating and the Borrower Rate, the higher or less risky the Prosper rating becomes, the lower the average Borrower Rate becomes. This is a very common act for loan companies

```python
plt.figure(figsize=(12,5));
sb.boxplot(data = df2, x = 'ProsperRating', y = 'BorrowerRate', color=base_color);
plt.xticks();
plt.title('Average, maximum and minimum Borrower Rates for Each Prosper Rating');
```

# Borrower Rate by Year and Income Range
The Average Borrower Rate becomes lower as the income range becomes higher, this also points to the fact that a borrowers Prosper Rating is likely to be higher as their income range increases. There seems to have been a spike in the borrower rates between 2011 and 2012 but the rates have been falling since then in most cases apart from unemplyed or zero income individuals who have experienced a higher spike in their borrower rates.

```python
def mean_poly(x, y, bins = 10, **kwargs):
    """ Custom adapted line plot code. """
    # set bin edges if none or int specified
    if type(bins) == int:
        bins = np.linspace(x.min(), x.max(), bins+1)
    bin_centers = (bin_edges[1:] + bin_edges[:-1]) / 2

    # compute counts
    data_bins = pd.cut(x, bins, right = False,
                       include_lowest = True)
    means = y.groupby(data_bins).mean()

    # create plot
    plt.errorbar(x = bin_centers, y = means, **kwargs)

bin_edges = np.arange(2005, 2014+1, 1)
g = sb.FacetGrid(data = df2, hue = 'IncomeRange', height = 5)
g.map(mean_poly, "Year", "BorrowerRate", bins = bin_edges)
g.set_ylabels('mean(BorrowerRate)')
g.add_legend()
plt.title("Average Borrower Rate per Year and Income Range");
plt.ylabel('Average Borrower Rate');
```

# Borrower Rate by Income Range and Home Owner Status

The average borrower rate is significantly lower for borrowers who own homes. This could be interpreted as the loan company predicting they are safer options and could also have a high value collateral asset in the case of defaulting on the loan. The borrower rate also follows the same theme as above in reducing as the income range increases but a point of interest is the average rate for borrowers with zero income who seem to have a lower rate than individuals earning between 1 and 49,999 Dollars. 

```python
sb.pointplot(data = df2, x = 'IncomeRange', y = 'BorrowerRate', hue = 'IsBorrowerHomeowner');
plt.xticks(rotation=30);
plt.title('Borrower Rate across Income Range and Home Owner Status');
plt.ylabel('Average Borrower Rate');
```
